## CH6 綜合練習：智慧醫療病患資料進階視覺化

### 情境說明
智慧醫療中心想以進階圖表同時觀察多個健康指標的分布與關聯。請運用本章學過的三種圖表（6.1 氣泡圖、6.2 分面圖 Facet Grid、6.3 成對變數矩陣圖 Pair Plot），呈現資料集中多變數之間的分布與關係。

### 資料說明
`datasets/smart_health_records.csv`（240 筆病患紀錄）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

### Try it!：CH6 綜合練習

**程式要求**
1. 讀取 `datasets/smart_health_records.csv`。
2. 氣泡圖（6.1）：以 `plt.scatter()` 呈現年齡（Age）與 BMI 的關係，氣泡大小代表住院天數、顏色依性別區分，並加上自訂圖例。
3. 分面圖（6.2）：以 `sns.relplot()` 依性別分面（`col="Gender"`），觀察年齡與血糖（Blood_Sugar）的關係。
4. 成對變數矩陣圖（6.3）：以 `sns.pairplot()` 呈現 Age、BMI、Blood_Sugar、Blood_Pressure 兩兩之間的關係，以 `hue` 依性別著色。
5. 每張圖皆需設定標題。
