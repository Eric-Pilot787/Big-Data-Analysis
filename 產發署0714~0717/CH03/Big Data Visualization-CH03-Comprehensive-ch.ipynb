## CH3 綜合練習：智慧醫療病患資料前處理

### 情境說明
智慧醫療中心從各院區彙整了一份病患紀錄，但原始資料含有缺失值、異常值、重複資料與型態問題。請運用本章 1~9 節學過的 pandas 前處理技巧，完成一條完整的資料清理流程。

### 資料說明
`dataset/smart_health_records_raw.csv`（245 筆，含 5 筆重複）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

已知資料問題：BMI 與 Blood_Sugar 含缺失值；Blood_Sugar 有大於 400 的異常值、Age 有大於 120 的異常值；Age 被存成浮點數；部分資料列完全重複。

### Try it!：CH3 綜合練習

**程式要求**
1. 讀取 `dataset/smart_health_records_raw.csv`，以 `head()`、`info()`、`describe()` 初探資料（對應 1 節）。
2. 以 `duplicated()` 檢查重複筆數，並用 `drop_duplicates()` 移除（對應 8 節）。
3. 以各欄位中位數填補 BMI 與 Blood_Sugar 的缺失值（對應 2 節）。
4. 移除異常值：Blood_Sugar 大於 400 或 Age 大於 120 的資料列（對應 2 節）。
5. 將 Age 轉換為整數型態（對應 3 節）。
6. 篩選出 60 歲以上的高齡病患並印出人數（對應 4 節）。
7. 依科別分組計算平均住院天數，並由高至低排序（對應 5 節）。
8. 建立科別對應樓層的小表格（Internal Medicine：3F、Surgery：5F、Cardiology：6F），以 `merge` 合併進病患資料（對應 6 節）。
9. 新增欄位 `Blood_Sugar_norm`：Blood_Sugar 的 min-max 正規化結果（對應 7 節），最後將清理後資料輸出成 `smart_health_records_cleaned.csv`（對應 9 節）。
