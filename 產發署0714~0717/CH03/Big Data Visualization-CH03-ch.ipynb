### **2. 缺失值與異常值處理**

**目的：**
- 檢查資料中的缺失值與可能存在的異常值，並採取適當的處理策略以提升資料品質。

**方法：**
- 使用 `isnull()` 檢查每個欄位的缺失值
- 透過 `dropna()` 刪除關鍵欄位的缺失值（例如診斷資訊）
- 利用 `fillna()` 以均值或中位數填補數值欄位缺失值

**範例程式：**

```python
# 檢查缺失值情形
print("\n各欄位缺失值數量：")
print(df.isnull().sum())

# 對數值欄位以均值/中位數填補缺失值
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].median())

# 對診斷欄位若缺失則刪除該筆記錄
df = df.dropna(subset=['Diagnosis'])

# 此外，利用描述統計或視覺化（此處僅示例）檢查異常值
print("\n更新後各欄位缺失值數量：")
print(df.isnull().sum())
```

### **5. 資料排序與分組統計**

**目的：**
- 對資料依據特定欄位進行排序，並分組計算各組別的統計數據，協助了解不同診斷群體的特徵。

**方法：**
- 使用 `sort_values()` 依據 AdmissionDate 進行排序
- 使用 `groupby()` 結合聚合函數（如 mean、sum、count）進行分組統計

**範例程式：**

```python
# 依據 AdmissionDate 進行排序（由舊至新）
df_sorted = df.sort_values(by='AdmissionDate', ascending=True)
print("\n依 AdmissionDate 排序後的資料：")
print(df_sorted.head())

# 以 Diagnosis 分組，計算各組別平均年齡與平均血壓
grouped = df.groupby('Diagnosis').agg({
    'Age': 'mean',
    'BloodPressure': 'mean'
}).reset_index()
print("\n以 Diagnosis 分組的統計結果：")
print(grouped)
```

### **6. 資料整合與轉換**

**目的：**
- 將來自不同來源的資料合併，或對同一資料集進行重塑，提升資料的結構性與可讀性。

**方法：**
- 使用 `merge()` 合併多個 DataFrame
- 使用 `pivot_table()` 重塑資料表結構
- 使用 `concat()` 合併資料列或欄

**範例程式：**

```python
# 假設利用同一資料集拆分出基本資料與檢查結果，再進行合併

# 提取基本資料子集（PatientID、Name、Age、Diagnosis）
df_basic = df[['PatientID', 'Name', 'Age', 'Diagnosis']]

# 提取檢查結果子集（PatientID、BloodPressure、AdmissionDate）
df_vitals = df[['PatientID', 'BloodPressure', 'AdmissionDate']]

# 以 PatientID 為關鍵欄位進行合併
df_merged = pd.merge(df_basic, df_vitals, on='PatientID', how='inner')
print("\n合併後的資料：")
print(df_merged.head())

# 利用 pivot_table 重塑資料，計算各診斷群組的平均血壓
pivot = pd.pivot_table(df_merged, values='BloodPressure', index='Diagnosis', aggfunc='mean')
print("\nPivot Table 統計各診斷群組平均血壓：")
print(pivot)
```

### **7. 資料標準化與正規化**

**目的：**
- 進行數值資料的標準化與正規化處理，以消除量綱差異，方便後續模型建立與分析。

**方法：**
- 使用 `sklearn.preprocessing.StandardScaler` 進行標準化
- 手動計算並實現 Min-Max 正規化

**範例程式：**

```python
from sklearn.preprocessing import StandardScaler

# 標準化 BloodPressure 欄位
scaler = StandardScaler()
df['BloodPressure_scaled'] = scaler.fit_transform(df[['BloodPressure']])
print("\n標準化後的 BloodPressure 前五筆：")
print(df[['BloodPressure', 'BloodPressure_scaled']].head())

# 手動進行 Age 欄位的 Min-Max 正規化
df['Age_norm'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
print("\n正規化後的 Age 前五筆：")
print(df[['Age', 'Age_norm']].head())
```

### **綜合範例程式**

下列程式碼整合前述各章節步驟，採用同一份醫療資料集 **medical_records.csv**，展示從資料讀取、清洗、轉換、篩選、排序、合併、標準化、去重到最後輸出的完整流程：

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# === 1. 數據讀取與初探 ===
# Kaggle API
import kagglehub

# Download latest version
path = kagglehub.dataset_download("f114118114/ch2-dataset")

print("Path to dataset files:", path)

from pathlib import Path

file_name = "medical_records.csv"
df = pd.read_csv(path +"/"+ file_name)
print("【初探資料】")
print(df.head())
print(df.info())
print(df.describe())

# === 2. 缺失值與異常值處理 ===
print("\n【缺失值檢查】")
print(df.isnull().sum())

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].median())
df = df.dropna(subset=['Diagnosis'])
print("缺失值處理後：")
print(df.isnull().sum())

# === 3. 資料型態轉換 ===
df['AdmissionDate'] = pd.to_datetime(df['AdmissionDate'])
df['Age'] = df['Age'].astype(int)
print("\n【資料型態轉換】")
print(df[['AdmissionDate', 'Age']].dtypes)

# === 4. 資料篩選與過濾 ===
df_filtered = df[(df['Age'] > 50) & (df['Diagnosis'] == 'Hypertension')]
print("\n【布林索引篩選】")
print(df_filtered.head())

df_filtered_query = df.query("Age > 50 and Diagnosis == 'Hypertension'")
print("【query() 篩選】")
print(df_filtered_query.head())

# === 5. 資料排序與分組統計 ===
df_sorted = df.sort_values(by='AdmissionDate', ascending=True)
print("\n【依 AdmissionDate 排序】")
print(df_sorted.head())

grouped = df.groupby('Diagnosis').agg({'Age': 'mean', 'BloodPressure': 'mean'}).reset_index()
print("\n【分組統計】")
print(grouped)

# === 6. 資料整合與轉換 ===
df_basic = df[['PatientID', 'Name', 'Age', 'Diagnosis']]
df_vitals = df[['PatientID', 'BloodPressure', 'AdmissionDate']]
df_merged = pd.merge(df_basic, df_vitals, on='PatientID', how='inner')
print("\n【合併資料】")
print(df_merged.head())

pivot = pd.pivot_table(df_merged, values='BloodPressure', index='Diagnosis', aggfunc='mean')
print("\n【Pivot Table】")
print(pivot)

# === 7. 資料標準化與正規化 ===
scaler = StandardScaler()
df['BloodPressure_scaled'] = scaler.fit_transform(df[['BloodPressure']])
print("\n【標準化】")
print(df[['BloodPressure', 'BloodPressure_scaled']].head())

df['Age_norm'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
print("【正規化】")
print(df[['Age', 'Age_norm']].head())

# === 8. 檢查與處理重複資料 ===
num_duplicates = df.duplicated().sum()
print(f"\n重複記錄數量： {num_duplicates}")
df = df.drop_duplicates()
print("去重後資料筆數：", len(df))

# === 9. 數據輸出與存檔 ===
df.to_csv("medical_records_cleaned.csv", index=False)
print("\n資料已成功輸出為 'medical_records_cleaned.csv'")
```
