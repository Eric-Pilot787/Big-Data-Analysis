## 2️⃣ 變數與資料型態

### 2.1 變數宣告與動態型別
Python 採用的是**動態型別**機制，這意味著：
1. **不需事先宣告型別**：當你第一次將資料指派給變數時，Python 會自動判斷並決定它的型別。
2. **隨時可變更**：你可以隨時將變數的資料型別變更。

**程式範例：**

```python
num = 10       # 整數型別
num = "十"     # 現在 num 為字串
```

### 2.2 常見資料型態及範例
為了應對不同的應用場景，Python 內建了多種基礎的資料型態：

* **整數 (int)：**
  包含正整數、負整數與零，沒有小數點。
  *如：`age = 25`、`temperature = -5`*
* **浮點數 (float)：**
  帶有小數點的數值，常用於需要精確計算的場合。
  *如：`pi = 3.14159`、`weight = 65.5`*
* **字串 (str)：**
  用來表示純文字。在 Python 中，只要被**單引號 (`'`)** 或**雙引號 (`"`)** 包起來的內容，都會被視為字串。
  *如：`name = "Alice"`、`message = 'Hello World'`*
* **布林值 (bool)：**
  只有兩種值：**`True`（真）** 與 **`False`（假）**。注意首字母必須大寫。通常用於條件判斷與邏輯控制。
  *如：`is_raining = True`、`is_passed = False`*
* **None 類型 (NoneType)：**
  這是一個特殊的型別，它只有一個值，就是 **`None`**。它代表「沒有值」、「空」或「尚未定義」。請注意，`None` 不等於數字的 `0`，也不等於空的字串 `""`，它就是單純的「什麼都沒有」。
  *如：`result = None`*

**程式範例：**

```python
a = 5       # 整數
b = 3.14    # 浮點數
greeting = "Hello, Python!"
is_active = True
result = None
print(a, b, greeting, is_active, result)
```

### 2.3 資料型態轉換
在程式開發中，資料的型態往往不是一成不變的。最常見的情境是：當我們透過 `input()` 接收使用者的輸入時，資料預設都是「字串 (`str`)」。

如果我們想拿它來做數學計算，就必須先將它轉換為「整數 (`int`)」或「浮點數 (`float`)」。

我們可以使用 Python 的內建函式來進行「強制轉型」：

* **`int()` 轉為整數：** 將字串或浮點數轉換為整數（若是浮點數，會直接無條件捨去小數點後的所有數字）。
* **`float()` 轉為浮點數：** 將整數或字串轉換為帶有小數點的數字。
* **`str()` 轉為字串：** 將數字或布林值轉換為純文字形式。這在需要將數字與文字合併輸出時非常實用。
* **`bool()` 轉為布林值：** 用來判斷資料的真偽。在 Python 中，數字 `0`、空字串 `""` 或空串列 `[]` 轉換後會是 `False`，其他有內容的資料轉換後通常為 `True`。

> **注意：** 轉換的內容必須合理。例如 `int("123")` 可以成功運作，但如果是 `int("abc")`，電腦不知道該怎麼把純文字變成數字，就會跳出錯誤程式中斷 (ValueError)。

**程式範例：**


```python
num_str = "123"
num_int = int(num_str)
num_float = float(num_str)
print(num_int, num_float)
```

### 2.4 基本運算
Python 提供了豐富的運算子，讓我們能像使用計算機一樣對變數與資料進行操作。主要可以分為以下四大類：

* **算術運算 (Arithmetic)：進行數學計算**
  * 除了基本的加 `+`、減 `-`、乘 `*`、除 `/` 之外。
  * 程式中還有非常實用的：取餘數 `%`（常用來判斷奇偶數）、取整數除法 `//`（除完後無條件捨去小數），以及次方計算 `**`。
* **賦值運算 (Assignment)：將資料存入變數**
  * 最基本的是單一個等號 `=`（將右邊的值存入左邊的變數）。
  * 進階用法如 `+=`、`-=`。例如 `a += 5` 是 `a = a + 5` 的縮寫，能讓程式碼更精簡。
* **比較運算 (Comparison)：用來比大小或比對內容**
  * 包含 `>`、`<`、`>=`、`<=`。
  > **注意：** 在程式中「判斷是否相等」是用兩個等號 `==`；而「不等於」則是 `!=`。比較的結果一定會產生布林值（`True` 或 `False`）。
* **邏輯運算 (Logical)：串聯或反轉多個條件**
  * **`and` (且)：** 當左右兩邊的條件都成立時，結果才為 `True`。
  * **`or` (或)：** 只要左右兩邊有一個條件成立，結果就是 `True`。
  * **`not` (非)：** 將結果反轉，`True` 變 `False`，`False` 變 `True`。

**程式範例：**

```python
x = 10
y = 3
print(x + y)   # 加法
print(x - y)   # 減法
print(x * y)   # 乘法
print(x / y)   # 除法
print(x // y)  # 整除
print(x % y)   # 取餘
print(x ** y)  # 指數

# 賦值、比較與邏輯運算
a = 5
a += 3
print(a)          # 輸出 8

print(5 == 5)     # True
print(5 != 3)     # True
print(5 > 2)      # True

print(True and False)  # False
print(not True)        # False
```

## 3️⃣ 條件判斷（if-else）

### 3.1 基本 if 判斷式
`if` 判斷式是程式邏輯的基礎，用來告訴電腦：「如果某個條件成立（結果為 `True`），就執行指定的程式區塊；如果不成立（結果為 `False`），就直接跳過。」

**語法重點與範例：**
在 Python 中寫條件判斷時，有兩個絕對不能忘記的規則：
1. 條件的結尾必須加上**冒號 (`:`)**。
2. 屬於該條件內的程式碼，必須進行**縮排（Indentation，通常是 4 個半形空白）**，Python 是以縮排來辨識哪些程式碼屬於同一個區塊。

**程式範例：**

```python
num = 10
if num > 0:
    print("正數")
```

### 3.2 if-else 與 if-elif-else 結構
當我們的程式不只有「做」或「不做」，而是有「非 A 即 B」或「多種不同情境」時，就需要加入 `else` 或 `elif` 來擴充判斷式。

* **`if-else` (雙向分支/二選一)：**
  如果 `if` 條件成立就做 A，否則（`else`）就做 B。
* **`if-elif-else` (多向分支/多選一)：**
  `elif` 是 "else if" 的縮寫。當有多個條件需要依序檢查時使用。

  程式會**由上往下**逐一檢查，**只要遇到第一個成立的條件，執行完該區塊後就會直接跳出整個判斷結構**，不會再理會後面的條件。

**程式範例：**

```python
score = 85
if score >= 90:
    print("優")
elif score >= 80:
    print("甲")
else:
    print("乙")
```

### 3.3 巢狀條件與單行 if 表達式
在撰寫較複雜的邏輯時，我們常常會遇到「前提條件」，或是希望能讓程式碼看起來更精簡。這時候就可以使用「巢狀條件」或「單行 if 表達式」。

**巢狀條件 (Nested Conditions)**

「巢狀條件」就是將一個 `if` 判斷式寫在另一個 `if` 判斷式的區塊裡面。

通常用於：**必須先滿足某個大前提，才能進一步做細部的判斷**。

**觀念重點：**
在使用巢狀條件時，**縮排（Indentation）非常重要**。

內層的 `if` 必須比外層的 `if` 多退一格縮排，這樣 Python 才能正確辨識邏輯的層次。

**程式範例：**

```python
# 巢狀條件
num = 7
if num > 0:
    if num % 2 == 0:
        print("正偶數")
    else:
        print("正奇數")

# 單行 if 表達式
result = "正數" if num > 0 else "非正數"
print(result)
```

## 4️⃣ 迴圈（for, while）

### 4.1 for 迴圈與 range() 函式
**`for` 迴圈**最常被用來「走訪（遍歷）」資料序列中的每一個元素，例如將列表中的項目逐一取出。

當我們需要程式「重複執行特定次數」時，通常會搭配內建的 `range()` 函式來產生一連串的數字序列。

**`range()` 函式的重點觀念：**
* 完整寫法為 `range(起始值, 結束值, 距離)`，如果只給兩個數字則是 `range(起始值, 結束值)`。
* **最大的新手陷阱**：產生的數字序列會包含「起始值」，但**絕對不包含「結束值」**（也就是常說的「包頭不包尾」）。

**程式範例：**

```python
for i in range(1, 6):
    print(i)
```

```python
for i in range(0,6,2):
    print(i)
```

### 4.2 while 迴圈
不同於 `for` 迴圈通常有明確的執行次數，`while` 迴圈是根據**「條件判斷」**來決定是否繼續重複。

只要條件的結果為 `True`，迴圈內的程式碼就會一直重複執行；直到條件變成 `False` 時，才會跳出迴圈。

**觀念重點與陷阱：**
1. 使用 `while` 迴圈時，**務必確保條件有變成 `False` 的一天**（通常會在迴圈內部更新變數的值）。
2. 如果忘記改變條件變數，程式就會陷入永遠無法停止的「無窮迴圈（Infinite Loop）」，導致整個程式卡死或崩潰。

**程式範例：**

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

### 4.3 break 與 continue 控制流程

在執行迴圈的過程中，我們有時會遇到某些特殊情況，需要提早結束迴圈，或是跳過某幾次不執行。這時就可以使用 Python 提供的兩個重要控制語句：

* **`break` (終止迴圈)：**
    * 一旦執行到 `break`，程式會立刻**中斷並跳出整個迴圈**，即使迴圈原本還有剩下的次數也不會再執行。
    * 通常搭配 `if` 判斷式，用在「找到目標了」或「發生錯誤」時提早結束。
* **`continue` (略過本次，繼續下次)：**
    * 執行到 `continue` 時，程式會**略過當次迴圈剩下的所有程式碼**，直接跳回迴圈的開頭，進入下一次的迴圈回合。
    * 這很像玩桌遊時的「暫停一回合後繼續」。

**程式範例：**

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)

for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i)
```

### 4.4 巢狀迴圈（九九乘法表）
巢狀迴圈（Nested Loops）」顧名思義，就是在一個迴圈的內部，再塞入另一個迴圈。

這在處理「二維資料」（例如表格、棋盤、矩陣）或是窮舉「多重組合」時非常實用。

**觀念重點：**
1. 外層迴圈每執行「1 次」，內層迴圈就會完整地執行「一整圈」。
2. 以九九乘法表為例，當外層的被乘數是 1 時，內層的乘數會跑完 1 到 9；接著外層變成 2，內層又會重新跑完 1 到 9。

**程式範例:**

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(f"{i} x {j} = {i*j}", end="\t")
    print()  # 換行
```

### Try it! 4：迴圈
「雞兔同籠」是一道經典的數學與程式邏輯題。已知籠子裡有`35`隻雞和兔子，共`94`隻腳，請撰寫一個程式，根據使用者輸入的「總頭數」與「總腳數」，計算出雞和兔子分別有幾隻。

**基本常識前提：**
* 雞有 1 個頭、2 隻腳。
* 兔子有 1 個頭、4 隻腳。

**程式要求**

1. 程式需請使用者輸入兩個整數：`總頭數` 與 `總腳數`。
2. 根據輸入的數值，計算出正確的雞與兔子數量，並印出結果。

```python
def solve_chicken_rabbit(total_heads, total_legs):
    for chickens in range(total_heads + 1):
        rabbits = total_heads - chickens
        if 2 * chickens + 4 * rabbits == total_legs:
            return chickens, rabbits
    return None, None

try:
    heads = int(input("請輸入總頭數："))
    legs = int(input("請輸入總腳數："))
except ValueError:
    print("請輸入整數。")
else:
    chickens, rabbits = solve_chicken_rabbit(heads, legs)
    if chickens is not None:
        print(f"雞：{chickens} 隻")
        print(f"兔：{rabbits} 隻")
    else:
        print("沒有符合條件的雞兔數量。")
```

> 這個題目也可以用遞迴完成，例如把「嘗試一個雞的數量，然後遞迴處理剩下的頭和腳」當作遞迴邏輯。但對於這種固定範圍的搜尋問題，使用迴圈會更簡潔、易懂。

## 5️⃣ 函式（Functions）

### 5.1 定義與呼叫函式
在程式開發的過程中，我們經常會遇到需要重複執行的特定邏輯。與其每次都重新撰寫相同的指令，不如將這些邏輯「封裝」起來，這就是函式（Function）的核心功用。

它就像是你自己發明的一句新指令。

* **定義函式：** 在 Python 中，我們使用 `def` 關鍵字（Define 的縮寫）來建立一個新的函式，並為它取一個具代表性的名稱，在名稱後方加上冒號與縮排，即可定義這個函式專屬的執行區塊。
* **呼叫函式：** 函式定義好之後並不會自動執行，當我們需要使用它時，必須寫出函式的名稱並加上小括號，這個動作稱為「呼叫（Call）」，程式便會跳去執行該函式內部的邏輯。

**程式範例:**

```python
def greet(name):
    """傳回問候語"""
    return f"Hello, {name}!"

print(greet("Alice"))
```

### 5.2 傳遞參數與回傳值

如果說函式是一台販賣機，那麼「參數」就是您投進去的零錢，而「回傳值」就是你要的飲品。

* **傳遞參數（Inputs）：** 定義函式時，可以在小括號內設定用來接收資料的變數，稱為「參數」。當我們呼叫函式時，可以將外部的實際資料傳遞進去，讓函式根據不同的資料產出對應的結果。
* **回傳值（Outputs）：** 函式執行完畢後，可以使用 `return` 關鍵字將運算後的最終結果「交還」給呼叫它的地方。一旦執行到 `return`，函式就會立刻結束。我們通常會將這個回傳的結果存入一個新的變數中，以便後續拿來做其他計算或輸出。

**程式範例：**

```python
def add(a, b):
    return a + b

result = add(10, 20)
print("和為：", result)
```

### 5.3 位置參數與預設參數
當一個函式需要接收多個參數時，Python 提供了彈性的機制來對應這些傳入的資料。

* **位置參數：** 這是最基本也最常用的傳遞方式，呼叫函式時，您所提供的資料會嚴格依照「由左至右」的順序，一一對應到函式定義時所設定的參數位置上，順序一旦放錯，程式的邏輯就會出錯。
* **預設參數：** 為了讓函式更具彈性，我們可以在定義函式時，直接在括號內為某些參數設定好「預設值」，未來在呼叫該函式時，如果使用者有提供該項資料，就會使用使用者提供的新資料；如果使用者忘記或省略不填，程式也不會報錯，而是會自動套用我們事先設定好的預設值。

**程式範例：**

```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))       # 預設 exponent 為 2，結果 9
print(power(3, 3))    # 結果 27
```

### 5.4 變數作用域
在程式中，變數並不是在任何地方都能隨意存取的，它們有自己的「應用範圍」與「可見度」，這就稱為「作用域」。

了解作用域能幫助我們避免變數名稱互相干擾（也就是所謂的命名衝突）。

* **全域變數 (Global Variable)：** 定義在所有函式之外的變數。它就像是公共資源，整個程式中的任何地方（包含各個函式內部）都可以「看見」並讀取它的值。
* **區域變數 (Local Variable)：** 定義在特定函式內部的變數。它就像是該函式的專屬私人物品，只有在函式執行期間才會存在；一旦函式執行完畢，這個變數就會自動被銷毀。外部的程式完全無法看見，也無法使用它。
* **同名衝突的處理機制：** 假設全域和區域剛好有一個同名的變數（例如都叫做 `x`），當我們在函式內部存取 `x` 時，Python 會「優先」看向並使用它自己的區域變數。這兩者雖然名字一樣，但在電腦記憶體中是完全獨立的兩個存在。因此，在函式內修改這個區域變數，絕對不會影響到外面的那個全域變數。

**程式範例：**

```python
x = 100  # 全域變數

def scope_test():
    x = 50  # 區域變數
    print("函式內 x =", x)

scope_test()
print("全域 x =", x)
```

### 5.5 Lambda 匿名函式
通常我們會使用 `def` 關鍵字來為函式正式命名與定義。但有時候，我們只需要一個邏輯非常簡單、甚至只會用到一次的小功能，特地為它想一個名字並定義一整段區塊，反而顯得多此一舉。這時就可以使用「Lambda 匿名函式」。

* **匿名與輕量化：** Lambda 函式顧名思義沒有名字，寫法極度精簡，通常強制規定只能寫成短短的「一行」程式碼。
* **運作邏輯：** 它是透過 `lambda` 關鍵字來啟動，接著寫下所需的接收參數，並在冒號後方直接寫出單一的數學或邏輯運算式。這行運算式的最終結果，會被系統自動當作回傳值交出，我們完全不需要也不能在裡面寫 `return`。
* **適用情境：** 這種函式最常被當作「拋棄式」的臨時邏輯。在進行進階的資料處理（例如針對串列進行排序、條件過濾、或是批次轉換）時，直接將 Lambda 函式當作條件塞入，能讓整體程式碼看起來更加簡潔、俐落且專業。

**程式範例：**

```python
square = lambda x: x * x
print("5 的平方為：", square(5))
```

## 6️⃣ 進階基礎：資料結構簡介

### 6.1 列表（List）
列表是 Python 中最基本且最常用的資料結構之一。您可以把它想像成一列擁有多個格子的火車，每個格子裡都可以裝載相同或完全不同種類的資料（例如同時裝載數字與字串）。

* **有序與索引機制：** 列表中的資料是「有嚴格順序」的。每一個被放入列表的元素，都會被分配到一個專屬的位置編號，我們稱為「索引（Index）」。
* **請務必記住一個關鍵：程式語言的索引是從 `0` 開始計算的！** 因此，列表中的第一個元素，其索引編號為 `0`，第二個為 `1`，依此類推。
* **高自由度的可變性：** 列表是一種「可變」的資料集合。這意味著您可以在程式執行期間，隨時透過索引去替換（修改）某個位置上的資料內容。
* **豐富的內建操作：** Python 針對列表提供了許多便利的操作工具：
  * **新增元素：** 您可以選擇將新資料直接附加到列表的最尾端（如同排隊排到最後面），也可以強行安插（插入）到列表中的任何指定位置。
  * **移除元素：** 您可以指定一個索引位置，將該位置的資料從列表中「彈出」並刪除；這個彈出的動作同時也能將該資料拿出來交給另一個變數使用。
* **列表推導式 (List Comprehension)：** 這是 Python 中非常具代表性的高階語法。當你需要透過某個規律的計算（例如將 1 到 5 的數字全部平方），或是加入特定條件來產生一連串資料並存入列表時，可以將原本需要寫成好幾行的迴圈，濃縮成極度精簡的一行語法。

**程式範例：**

```python
# 建立列表
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # 取得第一個元素

# 修改列表
fruits[1] = "blueberry"
print(fruits)

# 使用列表方法：append, insert, pop
fruits.append("dragonfruit")
fruits.insert(1, "kiwi")
removed_item = fruits.pop(2)
print("移除的項目：", removed_item)

# 列表推導式：產生 1~5 的平方數列表
squares = [x ** 2 for x in range(1, 6)]
print("平方數：", squares)
```

### 6.2 字典（Dictionary）
如果說列表像是一排有編號的置物櫃，那麼「字典」就像是我們真實世界中使用的查單字字典或電話簿。它不再依靠生硬的數字編號來尋找資料，而是依賴您自訂的「標籤」。

* **鍵值對 (Key-Value Pairs) 結構：** 字典內的每一筆資料，都是由一個「鍵（Key，可視為標籤或名稱）」與一個「值（Value，實際要儲存的資料內容）」緊密綁定在一起組成的，如:{Key:Value}
* **極速的資料查詢：** 字典的最大優勢在於「尋找資料的速度」。因為每筆資料都有明確且唯一的`Key`，當您想要讀取特定資料時，不需要像列表那樣從頭到尾一筆一筆慢慢找。只要告訴電腦目標資料的`Key`，它就能瞬間精準地把對應的`Value`提取出來，非常適合應用在需要頻繁查詢的大型資料庫情境。

**程式範例：**

```python
student = {"name": "Bob", "age": 21, "major": "Computer Science"}
print("名字：", student["name"])

for key, value in student.items():
    print(f"{key}: {value}")
```

### 6.3 集合（Set）
集合 (Set) 的概念與我們在數學課學到的集合非常相似。它有兩個最核心的特色：「絕對不重複」與「沒有固定順序」。

* **自動去重與無序性：** 如果將一堆包含重複項目的資料丟進集合中，它會自動過濾掉重複的部分，確保裡面的每一個元素都是獨一無二的。同時，因為集合內的資料是沒有特定順序的，所以我們無法像操作列表那樣，透過數字索引（編號）去抓取特定的元素。
* **強大的數學運算：** 集合真正的價值在於進行跨群組的邏輯比對。當我們有兩組不同的資料時，可以透過集合快速進行數學運算：
  * **交集：** 瞬間找出兩群資料中「共同擁有」的部分。
  * **聯集：** 將兩群資料合併，並自動剃除重複的項目。
  * **差集：** 找出只存在於 A 群，但不存在於 B 群的獨特項目。
  
**程式範例：**

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
print("交集：", set_a & set_b)
print("聯集：", set_a | set_b)
```

### 6.4 元組（Tuple）
元組 (Tuple) 在基本結構上與前面介紹過的「列表 (List)」非常相似，資料一樣有先後順序，我們也可以透過數字索引來讀取它。但它們之間有一個最關鍵的差異：元組是「不可變的 (Immutable)」。

* **資料的唯讀保險箱：** 一旦元組被建立起來，裡面的資料就如同刻在石頭上一樣，絕對無法再被新增、修改或刪除。這聽起來似乎限制很多，但這正是它的最大優勢。當我們需要儲存一些「絕對不應該被意外竄改」的固定資料（例如：地圖座標、螢幕解析度的寬度與高度、系統設定參數）時，使用元組是最安全的選擇。
* **解包特性 (Unpacking)：** 元組在 Python 中經常被用來進行「解包」操作。意思是我們可以將一個裝有多個資料的元組，一口氣將裡面的數值「一對一」地拆解並分配給相對應數量的變數，這讓資料的傳遞與指派變得極度簡潔。
  
**程式範例：**

```python
dimensions = (1920, 1080)
width, height = dimensions  # 解包元組
print("寬度：", width, "高度：", height)
```

### 6.5 二維陣列
在 Python 的原生語法中，其實並沒有獨立被稱為「二維陣列」的基礎資料型態。但我們經常需要處理表格化的數據，這時我們就會利用「列表中的列表 (List of Lists)」來完美模擬出二維陣列的結構。

* **結構化網格概念：** 您可以把二維陣列想像成一個 Excel 試算表或是一個棋盤。最外層的「大列表」代表著整個表格本身；而大列表裡面包著的每一個「小列表」，則代表著表格中的「一橫列 (Row)」。小列表裡面的資料，就是該列中的每一個儲存格元素。
* **常見應用場景：** 透過這種具備行與列 (Rows and Columns) 特性的多層次結構，我們能非常直覺且有效地處理複雜的數據形態，例如：數學上的多維矩陣運算、遊戲開發中的地圖網格生成，或是任何以表格形式呈現的資料庫數據。
  
**程式範例：**

```python
# 定義一個二維陣列（矩陣）
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# 存取特定元素（第2列第3欄，注意索引從0開始）
print("第2列第3欄的元素：", matrix[1][2])  # 輸出 6

# 逐列遍歷二維陣列並印出每一列
for row in matrix:
    print(row)
```

## 7️⃣ 檔案的輸入與輸出（以 CSV 檔案為例）

### 7.1 檔案操作的基本概念
利用 Python 的 `csv` 模組可輕鬆讀取與寫入 CSV 檔案，注意檔案路徑與開啟模式（讀取、寫入等）。

### 7.2 寫入 CSV 檔案
可使用 `csv.writer` 或 `csv.DictWriter` 將資料寫入 CSV 檔案。記得設定 `newline=""` 以避免在 Windows 系統中產生空行。

```python
import csv

data = [
    ["Name", "Age", "City"],
    ["Alice", 30, "Taipei"],
    ["Bob", 25, "Kaohsiung"],
    ["Charlie", 28, "Taichung"]
]

with open("data.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(data)

print("CSV 檔案 data.csv 已建立")
```

### 7.3 讀取 CSV 檔案
可使用 `csv.reader` 或 `csv.DictReader` 方法讀取 CSV 檔案，並以列表或字典的形式存取資料。建議使用 `with` 陳述式確保檔案正確關閉。

```python
import csv

# 使用 csv.reader 讀取 CSV 檔案
with open("data.csv", "r", newline="", encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)

# 使用 csv.DictReader 以字典形式讀取
with open("data.csv", "r", newline="", encoding="utf-8") as csvfile:
    dict_reader = csv.DictReader(csvfile)
    for row in dict_reader:
        print(row)
```

### Try it! 7：CSV 檔案操作
撰寫一個程式，讀取一個 CSV 檔案（假設每一列皆為數字資料），計算每一列的平均值，並將所有平均值寫入新的 CSV 檔案中。

```python
import csv

averages = []
# 讀取 CSV 檔案並計算每列的平均值
with open("data.csv", "r", newline="", encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # 假設第一列為標題
    for row in reader:
        numbers = [float(num) for num in row]
        avg = sum(numbers) / len(numbers)
        averages.append([avg])

# 寫入新的 CSV 檔案，包含平均值
with open("averages.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Average"])
    writer.writerows(averages)

print("每列平均值已寫入 averages.csv")
```
