## CH2 綜合練習：智慧醫療病患資料分析

### 情境說明
你是智慧醫療中心的資料助理，需要在不使用任何資料分析套件的情況下，僅以本章學過的 Python 基本語法（變數與資料型態、條件判斷、迴圈、函式、資料結構、CSV 檔案讀寫）完成一份病患資料的基本統計。

### 資料說明
`smart_health_records.csv`（240 筆病患紀錄）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

### Try it!：CH2 綜合練習

**程式要求**
1. 使用 `csv.DictReader` 讀取 `smart_health_records.csv`，將每筆資料存入 list（對應 6、7 節：資料結構、檔案輸入）。
2. 將每筆資料的 Age 轉為 int、BMI 與 Blood_Sugar 轉為 float（對應 2 節：資料型態轉換）。
3. 定義函式 `bmi_category(bmi)`，以 if-elif-else 回傳「過輕（<18.5）／正常（18.5~24）／過重（24~27）／肥胖（>=27）」（對應 3、5 節：條件判斷、函式）。
4. 以 for 迴圈搭配 dict 統計各 BMI 分類的人數，並計算全體平均血糖，以 print 輸出結果（對應 4、6 節：迴圈、資料結構）。
5. 以 while 迴圈依序找出前 5 位血糖超過 140 的病患編號並輸出（對應 4 節：迴圈）。
6. 將各 BMI 分類人數寫入新檔案 `bmi_summary.csv`（欄位：BMI_Category、Count），並印出完成訊息（對應 7 節：檔案輸出）。
