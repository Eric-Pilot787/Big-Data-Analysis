## CH5 綜合練習：智慧醫療病患資料多變數視覺化

### 情境說明
智慧醫療中心想進一步了解病患各項健康指標之間的關係。請運用本章學過的四種圖表（5.1 散佈圖、5.2 群組長條圖、5.3 熱力圖、5.4 氣泡圖），呈現資料集中多變數之間的關係與比較。

### 資料說明
`datasets/smart_health_records.csv`（240 筆病患紀錄）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

### Try it!：CH5 綜合練習

**程式要求**
1. 讀取 `datasets/smart_health_records.csv`。
2. 散佈圖（5.1）：以 `sns.scatterplot()` 呈現年齡（Age）與血糖（Blood_Sugar）的關係，並以 `hue` 依性別著色。
3. 群組長條圖（5.2）：以 `sns.barplot()` 比較各科別、不同性別的平均住院天數（x 為 Department、hue 為 Gender）。
4. 熱力圖（5.3）：計算 Age、BMI、Blood_Sugar、Blood_Pressure、Hospital_Stay_Days 的相關係數矩陣，以 `sns.heatmap()` 呈現並標註數值。
5. 氣泡圖（5.4）：以 `plt.scatter()` 呈現年齡與 BMI 的關係，氣泡大小代表住院天數、顏色依性別區分。
6. 每張圖皆需設定標題與軸標籤。
