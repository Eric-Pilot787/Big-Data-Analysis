## CH7 綜合練習：智慧醫療病患資料互動式視覺化

### 情境說明
智慧醫療中心希望以互動式圖表製作病患資料儀表板。請運用本章 7.3 節學過的七種 Plotly 圖表（散點圖、長條圖、折線圖、面積圖、箱型圖、小提琴圖、直方圖＋密度圖），呈現資料集中單變數或多變數的分布、關係與趨勢。

### 資料說明
`datasets/smart_health_records.csv`（240 筆病患紀錄）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

### Try it!：CH7 綜合練習

**程式要求**
1. 讀取 `datasets/smart_health_records.csv`。
2. 散點圖（7.3.1）：`px.scatter` 呈現年齡與血糖的關係，依性別著色並加上 `trendline="ols"` 趨勢線。
3. 長條圖（7.3.2）：先以 `groupby` 計算各科別平均血糖，再以 `px.bar` 呈現。
4. 折線圖（7.3.3）：將年齡以每 10 歲分組（20-29、30-39、...、80-89），以 `px.line` 呈現各年齡組的平均血糖趨勢。
5. 面積圖（7.3.4）：以 `px.area` 呈現隨年齡累積的住院總天數。
6. 箱型圖（7.3.5）：`px.box` 比較各科別血糖分布。
7. 小提琴圖（7.3.6）：`px.violin` 比較不同性別的 BMI 分布（`box=True`）。
8. 直方圖＋密度圖（7.3.7）：以 `create_distplot` 依性別分組呈現血糖分布（含下方 rug）。
9. 每張圖皆需以中文設定標題與軸標籤。
