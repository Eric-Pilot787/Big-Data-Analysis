## 4.2 直方圖（Histogram）

### 1. 定義  
直方圖用於顯示**連續型數值變數的分佈情況**，透過將數值範圍分成一組組區間（bin），並以長條高度表示每個區間內的資料數量。


### 2. 範例圖表與說明  

  <img src="https://raw.githubusercontent.com/jacklin92/ch4/main/Histgram-00.png" width="400">


**觀察與解釋**：  
此直方圖描繪顧客年齡的分布，大多數年齡集中在 30–40 歲區間，呈現接近常態分布。可協助業者定位目標客群及設計對應行銷活動。


### 3. 適用情況  
- 連續數據分佈觀察（如年齡、薪資、溫度）  
- 探索數據偏態、集中趨勢與離群值


### 4. 優缺點  
- **優點**：  
  - 適合視覺化大量連續數據的分佈  
  - 能揭示數據的集中、偏態與變異性  
- **缺點**：  
  - 資料分箱數（bin）設定影響圖表解讀  
  - 不適用於類別資料

#### 💻 實作目標

透過 Seaborn 套件繪製「連續數值資料的直方圖」，並練習以下技能：

- 使用 `sns.histplot()` 畫出分布視覺化
- 設定分箱數量（`bins`）
- 加入 KDE 曲線（趨勢線）
- 自訂顏色、邊框與圖表樣式


```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
path = '/content/drive/MyDrive/Colab Notebooks/CH4_dataset/datasets'
img_path = '/content/drive/MyDrive/Colab Notebooks/CH4_dataset/images'
```

```python
# 匯入所需套件
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 讀取病人年齡資料（CSV 檔）
from pathlib import Path

data_file = "medical_patient_ages.csv"
df = pd.read_csv(path+"/"+data_file)

# 檢查資料內容
print(df.head())

# 設定 seaborn 主題樣式
sns.set(style="whitegrid")

# 建立圖表區域，設定圖表尺寸
plt.figure(figsize=(8, 5))

# 使用 seaborn 繪製 Histogram（直方圖）
sns.histplot(
    data=df,
    x="Age",             # 要視覺化的欄位
    bins=15,             # 分成 15 個年齡區間
    kde=True,            # 加上平滑曲線 (Kernel Density Estimation)
    color="skyblue",     # 長條顏色（color 可用名稱或 HEX）
    edgecolor="black"    # 長條邊框顏色
)

# color 常見設定值：
# 'skyblue', 'salmon', 'lightgreen', 'gold', 'violet', '#1f77b4', '#FF5733'

# edgecolor 常見設定值：
# 'black', 'white', 'gray', '#444444', '#CCCCCC'

# 圖表標題與軸標籤設定
plt.title("Distribution of Patient Ages", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)

# 自動調整排版
plt.tight_layout()

# 顯示圖表
plt.show()
```
