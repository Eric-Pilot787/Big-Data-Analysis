## 4.3 圓餅圖（Pie Chart）

### 1. 定義  
圓餅圖利用整體圓形被切成扇形的方式，呈現各類別在總體中所佔比例，適合用於表達百分比概念。


### 2. 範例圖表與說明
<div align="center">  
  <img src="https://raw.githubusercontent.com/jacklin92/ch4/main/Pie-00.png" width="400">
</div>

**觀察與解釋**：  
圖中呈現不同付款方式的使用比例。Credit Card 占 45%，為最常見付款方式；Cash 僅占 10%。適合用來分析顧客偏好的支付行為，並優化結帳流程。


### 3. 適用情況  
- 比較多組數據的分布情形與差異
- 快速辨識並標示出資料中的離群值（異常值）
- 呈現數據的集中趨勢與分散程度


### 4. 優缺點  
- **優點**：  
  - 清楚呈現資料的五數總結（中位數、四分位數、極值）與離群值
  - 非常適合並排比較多種類別的數據，節省版面且簡潔
  - 不受資料量大小影響，能有效概括大量數據
- **缺點**：  
  - 無法顯示資料的內部真實分布形狀（例如可能會隱藏雙峰分布）
  - 對於缺乏統計基礎的大眾而言，較難第一時間直觀理解
  - 無法直接從圖表中看出各組資料的確切樣本數（除非額外標示）

### 6. 實作範例：醫學領域 - 各類疾病住院比重分析（Pie Chart）

#### 📁 資料說明

本範例模擬某醫療機構中因各種疾病類型而住院的病人數量。

| Disease_Category   | Hospitalized_Patients |
|--------------------|------------------------|
| Cardiovascular     | 120                    |
| Respiratory        | 80                     |
| Neurological       | 60                     |
| Gastrointestinal   | 40                     |
| Other              | 30                     |

📥 [下載 CSV 檔案：`hospitalized_by_disease.csv`](sandbox:/mnt/data/hospitalized_by_disease.csv)


#### 💻 實作目標

透過 Matplotlib 套件繪製「分類資料所佔比例的圓餅圖」，並練習以下技能：

- 使用 `plt.pie()` 畫出資料比例分布  
- 加入標籤與百分比文字  
- 設定顏色、突顯區塊與起始角度  
- 製作美觀且易讀的圓餅圖報表


````python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
path = '/content/drive/MyDrive/Colab Notebooks/CH4_dataset/datasets'
img_path = '/content/drive/MyDrive/Colab Notebooks/CH4_dataset/images'
````

````python
# 匯入所需套件
import pandas as pd
import matplotlib.pyplot as plt

# 讀取資料集
from pathlib import Path

data_file = "hospitalized_by_disease.csv"
df = pd.read_csv(path+"/"+data_file)

# 準備圖表用資料
labels = df["Disease_Category"]                   # 各區塊標籤
sizes = df["Hospitalized_Patients"]               # 各類別對應數量

# 設定自訂色彩（color 可用名稱或 HEX 色碼）
colors = ["#FF9999", "#66B3FF", "#99FF99", "#FFD700", "#C2C2F0"]

# explode：突顯最大類別（此例為 Cardiovascular）
explode = [0.05 if i == max(sizes) else 0 for i in sizes]

# 建立圓餅圖
plt.figure(figsize=(7, 7))  # 圖表大小（寬 x 高）

plt.pie(
    sizes,
    labels=labels,              # 顯示標籤名稱
    colors=colors,              # 區塊顏色
    explode=explode,            # 突顯最大類別
    autopct="%1.1f%%",          # 顯示百分比，保留一位小數
    startangle=140,             # 起始角度
    shadow=True                 # 是否加入陰影
)

# 設定圖表標題
plt.title("Hospitalized Patients by Disease Category", fontsize=14)

# 使圓形不變形
plt.axis("equal")

# 自動調整版面
plt.tight_layout()

# 顯示圖表
plt.show()
````
