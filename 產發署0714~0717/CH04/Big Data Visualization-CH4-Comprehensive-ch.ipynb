## CH4 綜合練習：智慧醫療病患資料視覺化

### 情境說明
智慧醫療中心希望以一份病患資料集製作視覺化報告。請運用本章學過的四種圖表（4.1 長條圖、4.2 直方圖、4.3 圓餅圖、4.4 箱型圖），分別呈現資料集中單變數或多變數的分布與比較。

### 資料說明
`datasets/smart_health_records.csv`（240 筆病患紀錄）

| 欄位 | 說明 |
|------|------|
| Patient_ID | 病患編號 |
| Gender | 性別（Male / Female） |
| Department | 科別（Internal Medicine / Surgery / Cardiology） |
| Age | 年齡 |
| BMI | 身體質量指數 |
| Blood_Sugar | 血糖值（mg/dL） |
| Blood_Pressure | 收縮壓（mmHg） |
| Hospital_Stay_Days | 住院天數 |

### Try it!：CH4 綜合練習

**程式要求**
1. 讀取 `datasets/smart_health_records.csv`。
2. 長條圖（4.1）：統計各科別（Department）病患人數，以 `plt.bar()` 呈現類別分布。
3. 直方圖（4.2）：以 `plt.hist()` 呈現全體病患年齡（Age）的分布，`bins=14`。
4. 圓餅圖（4.3）：以 `plt.pie()` 呈現各科別病患人數占比，顯示百分比。
5. 箱型圖（4.4）：以 `plt.boxplot()` 比較三個科別血糖值（Blood_Sugar）的分布。
6. 每張圖皆需設定標題與軸標籤。
